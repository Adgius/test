```python
import geopandas as gpd
import pandas as pd
```

# RUSSIA

```python
df = gpd.read_file('russia_geojson_wgs84.geojson')
```

```python
df = df[['name', 'geometry']]
```

```python
df.plot()
```

# KZ

```python
import requests as re

kz = re.get('https://raw.githubusercontent.com/open-data-kazakhstan/geo-boundaries-kz/master/data/geojson/kz_1.json')
```

```python
with open('kz.geojson', 'w') as f:
    f.writelines(kz.content.decode('windows-1251'))
```

```python
kz_data = gpd.read_file('kz.geojson')
```

```python
kz_data = kz_data[['NAME_1', 'geometry']].rename(columns={'NAME_1': 'name'})
```

```python
kz_data.plot()
```

# BY

```python
by = re.get('https://github.com/wmgeolab/geoBoundaries/raw/905b0ba/releaseData/gbOpen/BLR/ADM1/geoBoundaries-BLR-ADM1_simplified.geojson')
```

```python
with open('by.geojson', 'w') as f:
    f.writelines(by.content.decode('windows-1251'))
```

```python
by_data = gpd.read_file('by.geojson')
```

```python
by_data = by_data[['shapeName', 'geometry']].rename(columns={'shapeName': 'name'})
```

```python
by_data.plot()
```

```python
with open('all_countries.geojson', 'w') as f:
    f.writelines(gpd.GeoDataFrame(pd.concat([df, kz_data, by_data], ignore_index=False)).to_json())
```
